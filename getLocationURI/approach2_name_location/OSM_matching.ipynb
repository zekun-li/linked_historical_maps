```python
import json

with open('map_phrase_USGS(GT).json') as f:
    history = json.load(f)
```

```python
history
```

```python
def getlower(osm):
    new = set()
    for word in osm:
    #     print(word.lower())
        new = new.union({word.lower()})
    return new
```

```python
matching_dic = dict()
num_dic = dict()
not_matching_dic = dict()
for mapname in history:    
    try:
        df = pd.read_csv(osm_data / "{}.tsv".format(mapname), sep='\t')
        osm = getlower(df.name)
        
        history_num = len(history[mapname])
        osm_num = len(osm)
        
        tmp_list = list()
        tmp_list2 = list() # not matching word
        for word in history[mapname]:
            if word in osm:
                tmp_list.append(word)
            else: 
                tmp_list2.append(word)
        matching_dic[mapname] = tmp_list
        not_matching_dic[mapname] = tmp_list2
        matching_num = len(tmp_list)
        
        num_dic[mapname] = [osm_num, history_num, matching_num, len(tmp_list2)]
        if history_num != matching_num+ len(tmp_list2):
            print(mapname)
        
    except:
        continue
```

```python
url_dic = dict()
for mapname in matching_dic:
    phrase_dic = dict()
    df = pd.read_csv(osm_data /"{}.tsv".format(mapname), sep='\t')
    for phrase in matching_dic[mapname]:        
        for i, name in enumerate(df.name):
            if name.lower() == phrase:
                url = "https://www.openstreetmap.org/"+str(df.iloc[i].osm_type)+"/"+str(df.iloc[i].osm_id)             
               
                if name in phrase_dic.keys():
                    phrase_dic[name].append(url)
                    if type(df.iloc[i].wikidata) == str:
                        wiki_url = 'http://www.wikidata.org/entity/'+df.iloc[i].wikidata
                        phrase_dic[name].append(wiki_url)                   
                else:
                    phrase_dic[name] = [url]
                    if type(df.iloc[i].wikidata) == str:
                        wiki_url = 'http://www.wikidata.org/entity/'+df.iloc[i].wikidata
                        phrase_dic[name].append(wiki_url) 
                    
    url_dic[mapname] = phrase_dic
            
```

```python
abbreviation_dic = {'st': "street", 'ave': "avenue", 'cr':'creek','cr.':'creek', 
                    'mt': 'mountain', 'mt.': 'mountain', 'mtn':'mountain','mtn.':'mountain',  
                    'pk': 'peak', 'blvd': 'boulevard'}
```

```python
# update url_dic
not_matching_dic2 = dict()
history_osm_dic = dict()  # abbreviation matching dictionary 

for mapname in not_matching_dic:
    phrase_dic = url_dic[mapname]
    df = pd.read_csv(osm_data /"{}.tsv".format(mapname), sep='\t')
    names = list(df.name)
    not_matching_phrase_list = list()
    for phrases in not_matching_dic[mapname]:
        phrase_list = phrases.split()
        phrase_list2 = list()
        flag = False
        for phrase in phrase_list:
            if phrase in abbreviation_dic.keys():
                phrase_list2.append(abbreviation_dic[phrase])  # replace abbreviation 'st', 'ave' into 'street', 'avenue'
                flag = True
            else:
                phrase_list2.append(phrase)
        
        if flag:
            set1 = set(phrase_list2)  # abbreviations are replaced in set1         
            flag2 = False
            for i, name in enumerate(names):
                set2 = set([n.lower() for n in name.split()])                
                if set1 == set2:
                    flag2 = True
                    
                    if name in history_osm_dic.keys():
                        history_osm_dic[name] = history_osm_dic[name].union([phrases])
                    else:
                        history_osm_dic[name] = set([phrases]) 
                    
                    
                    url = "https://www.openstreetmap.org/"+str(df.iloc[i].osm_type)+"/"+str(df.iloc[i].osm_id)
                    
                      
                    if name in phrase_dic.keys():
                        phrase_dic[name].append(url)
                        if type(df.iloc[i].wikidata) == str:
                            wiki_url = 'http://www.wikidata.org/entity/'+df.iloc[i].wikidata
                            phrase_dic[name].append(wiki_url)                   
                    else:
                        phrase_dic[name] = [url]
                        if type(df.iloc[i].wikidata) == str:
                            wiki_url = 'http://www.wikidata.org/entity/'+df.iloc[i].wikidata
                            phrase_dic[name].append(wiki_url)
                        
            if flag2 == False:        
                not_matching_phrase_list.append(phrases)
                                    
        else:
            not_matching_phrase_list.append(phrases)
                        
    not_matching_dic2[mapname] = not_matching_phrase_list
        
    url_dic[mapname] = phrase_dic                
            
            
            
```

```python
pair_dic = dict()
for pair in pair_list:
    pair_dic[pair[0]] =pair[1]
pair_dic
```

```python
# update url_dic with pair_list
not_matching_dic3 = dict()
# history_osm_dic = dict()  # abbreviation matching dictionary 

for mapname in not_matching_dic2:
    phrase_dic = url_dic[mapname]
    df = pd.read_csv(osm_data /"{}.tsv".format(mapname), sep='\t')
    names = list(df.name)
    not_matching_phrase_list = list()
    for phrase in not_matching_dic2[mapname]:
        if phrase in pair_dic.keys():
            osm_name = pair_dic[phrase]           
     
            for i, name in enumerate(names):
                if name.lower() == osm_name:
                    url = "https://www.openstreetmap.org/"+str(df.iloc[i].osm_type)+"/"+str(df.iloc[i].osm_id)             
                    
                    if name in phrase_dic.keys():
                        phrase_dic[name].append(url)
                        if type(df.iloc[i].wikidata) == str:
                            wiki_url = 'http://www.wikidata.org/entity/'+df.iloc[i].wikidata
                            phrase_dic[name].append(wiki_url)                   
                    else:
                        phrase_dic[name] = [url]
                        if type(df.iloc[i].wikidata) == str:
                            wiki_url = 'http://www.wikidata.org/entity/'+df.iloc[i].wikidata
                            phrase_dic[name].append(wiki_url)
                            
                    if name in history_osm_dic.keys():
                        history_osm_dic[name] = history_osm_dic[name].union([phrase])
                    else:
                        history_osm_dic[name] = set([phrase]) 
                            
        else:
            not_matching_phrase_list.append(phrase)            

                      
    not_matching_dic3[mapname] = not_matching_phrase_list        
    url_dic[mapname] = phrase_dic      
 
```
