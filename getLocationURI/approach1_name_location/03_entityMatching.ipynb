```python
"""
URIs
- Case-senstive Exact matching phrases 
- withing the location boundaris
- node, way (no relation)
"""
import pandas as pd
location_info = pd.read_csv("Historical Map Metadata - USGS.csv")
```

```python
# location_info['Geographic coordinates <lat, lng>']
location_info
```

```python
def getCoordinate(location):
    lat_list = set()
    long_list = set()
    for pair in location.split(';'):
        lat_list.add(float(pair.split(',')[0]))
        long_list.add(float(pair.split(',')[1]))
        
    return lat_list, long_list

loc_dic = dict()
for i in range(len(location_info)):
    lat_list, long_list = getCoordinate(location_info.iloc[i, 1])
    loc_dic[location_info.iloc[i, 0]+'.jpg'] = {'lat':lat_list, "long":long_list}
```

```python
loc_dic
```

```python
import pickle
from pathlib import Path

intermediate_output = Path("intermediate_output")

with open(intermediate_output / 'name_loc_result.pkl', 'rb') as handle:
    uri_dic = pickle.load(handle)
```

```python
# example
uri_dic['USGS-15-CA-brawley-e1957-s1957-p1961.jpg']['El Centro']
```

```python
new_uri_dic = dict()
# empty_phrase_dic = dict()

for map_name in uri_dic:
    tmp_dic = dict()
    
    # boundary of map from metadata
    gt_lat = loc_dic[map_name]['lat']
    gt_long = loc_dic[map_name]['long']
   
    for phrase in uri_dic[map_name]:
        tmp_uri_dic = dict()

        for uri in uri_dic[map_name][phrase]:

            if {'lat', 'long'}.issubset(uri_dic[map_name][phrase][uri].keys()):            
                
                lat = float(uri_dic[map_name][phrase][uri]['lat'])
                long = float(uri_dic[map_name][phrase][uri]['long'])
           
                if (lat >= min(gt_lat)) and (lat <= max(gt_lat)) and (long >= min(gt_long)) and (long <= max(gt_long)):
                    tmp_uri_dic[uri] = uri_dic[map_name][phrase][uri]
                
            elif {'longlat'}.issubset(uri_dic[map_name][phrase][uri].keys()):
                loc_info = uri_dic[map_name][phrase][uri]['longlat']
#                 print(uri_dic[map_name][phrase][uri])
                if loc_info.startswith( 'LINESTRING' ):                     
                    for ele in loc_info[11:-1].split(','):
                        pair = ele.split(' ')
                        lat = float(pair[1])
                        long = float(pair[0])
                        
                        if (lat >= min(gt_lat)) and (lat <= max(gt_lat)) and (long >= min(gt_long)) and (long <= max(gt_long)):
                            #uri_list.append(uri_dic[map_name][phrase][uri])
                            tmp_uri_dic[uri] = uri_dic[map_name][phrase][uri]
                            break
                            
                else:
                    pair =  loc_info[6:-1].split(' ')
                    lat = float(pair[1])
                    long = float(pair[0])

                    if (lat >= min(gt_lat)) and (lat <= max(gt_lat)) and (long >= min(gt_long)) and (long <= max(gt_long)):
                        tmp_uri_dic[uri] = uri_dic[map_name][phrase][uri]
#                         uri_list.append(uri_dic[map_name][phrase][uri])
                        
            else:
                try:
                    lat = float(uri_dic[map_name][phrase][uri]['lat_geo'])
                    long = float(uri_dic[map_name][phrase][uri]['long_geo'])
                    if (lat >= min(gt_lat)) and (lat <= max(gt_lat)) and (long >= min(gt_long)) and (long <= max(gt_long)):
                        tmp_uri_dic[uri] = uri_dic[map_name][phrase][uri]
#                         uri_list.append(uri_dic[map_name][phrase][uri])
                except:
                    continue
        
        if tmp_uri_dic != {}:
            tmp_dic[phrase] = tmp_uri_dic
    
    new_uri_dic[map_name] = tmp_dic
```

```python
import json

with open('map_phrase_USGS(GT).json', 'r') as f: # extractPhrase_USGS(GT).ipynb in make RDF folder
    phrase_dic = json.load(f)
    
# make "black mountain" to "Black Mountain"
new_dic = dict()
for key in phrase_dic:
    new_phrase = list()
    for phrase in phrase_dic[key]:
        string = ""
        for voc in phrase.split():
            string += voc.capitalize() + " "
        new_phrase.append(string[:-1])
    new_dic[key] = new_phrase
```

```python
new_dic
```

```python
empty_uri_dic = dict()
for key in new_dic:
     empty_uri_dic[key] = list(set(new_dic[key]) - set(new_uri_dic[key].keys()))
```

```python
output = Path("output")

with open(output / 'uri_exactmatching.pkl', 'wb') as handle:
    pickle.dump(new_uri_dic, handle)
    
with open(output / 'empty_uri_phrases.pkl', 'wb') as handle:
    pickle.dump(empty_uri_dic, handle)
```

```python
number_matching = dict()
for map_name in new_uri_dic:
    number_matching[map_name] = {"matching": len(new_uri_dic[map_name].keys())}
```

```python
for map_name in number_matching:
    number_matching[map_name].update({"total": len(phrase_dic[map_name])})
```

```python
for map_name in number_matching:
    number_matching[map_name].update({"not matching": len(empty_uri_dic[map_name])})
    
    if len(empty_uri_dic[map_name]) != number_matching[map_name]['total']-number_matching[map_name]['matching']:
        print(map_name)
```
