```python
import pandas as pd
location_info = pd.read_csv("Historical Map Metadata - USGS.csv")
```

```python
# location_info['Geographic coordinates <lat, lng>']
location_info
```

```python
def getCoordinate(location):
    lat_list = set()
    long_list = set()
    for pair in location.split(';'):
        lat_list.add(float(pair.split(',')[0]))
        long_list.add(float(pair.split(',')[1]))
        
    return lat_list, long_list

loc_dic = dict()
for i in range(len(location_info)):
    lat_list, long_list = getCoordinate(location_info.iloc[i, 1])
    loc_dic[location_info.iloc[i, 0]+'.jpg'] = {'lat':lat_list, "long":long_list}
```

```python
import pickle

with open('sparql_usgs_result1.pkl', 'rb') as handle:
    uri_dic = pickle.load(handle)
```

```python
# example
uri_dic['USGS-15-CA-brawley-e1957-s1957-p1961.jpg']['El Centro']
```

```python
new_uri_dic = dict()
# empty_phrase_dic = dict()

for key in uri_dic:
    tmp_dic = dict()
    for phrase in uri_dic[key]:
        loc_list = list()
        for loc in uri_dic[key][phrase]:
            lat = float(loc[1])
            long = float(loc[2])

            if (lat >= min(loc_dic[key]['lat'])) and (lat <= max(loc_dic[key]['lat'])) and (long >= min(loc_dic[key]['long'])) and (long <= max(loc_dic[key]['long'])):
                loc_list.append(loc)
        
        if loc_list != []:
            tmp_dic[phrase] = loc_list
    
    new_uri_dic[key] = tmp_dic
```

```python
new_uri_dic2 = dict()
for key in uri_dic2:
    tmp_dic = dict()
    for phrase in uri_dic2[key]:
        loc_list = list()
        for loc in uri_dic2[key][phrase]:
            lat = float(loc[1])
            long = float(loc[2])

            if (lat >= min(loc_dic[key]['lat'])) and (lat <= max(loc_dic[key]['lat'])) and (long >= min(loc_dic[key]['long'])) and (long <= max(loc_dic[key]['long'])):
                loc_list.append(loc)
        
        if loc_list != []:
            tmp_dic[phrase] = loc_list
    new_uri_dic2[key] = tmp_dic
```

```python
import json

with open('map_phrase_USGS(GT).json', 'r') as f: # extractPhrase_USGS(GT).ipynb in make RDF folder
    phrase_dic = json.load(f)
    
# make "black mountain" to "Black Mountain"
new_dic = dict()
for key in phrase_dic:
    new_phrase = list()
    for phrase in phrase_dic[key]:
        string = ""
        for voc in phrase.split():
            string += voc.capitalize() + " "
        new_phrase.append(string[:-1])
    new_dic[key] = new_phrase
```

```python
new_dic
```

```python
for key in new_dic:
    for phrase in new_uri_dic2[key].keys():
        if phrase not in new_uri_dic[key].keys():
            new_uri_dic[key].update({phrase: new_uri_dic2[key][phrase]})
```

```python
new_uri_dic
```

```python
empty_uri_dic = dict()
for key in new_dic:
     empty_uri_dic[key] = list(set(new_dic[key]) - set(new_uri_dic[key].keys()).union(new_uri_dic2[key].keys()))
```

```python
with open('uri_exactmatching.pkl', 'wb') as handle:
    pickle.dump(new_uri_dic, handle)
    
with open('empty_uri_phrases.pkl', 'wb') as handle:
    pickle.dump(empty_uri_dic, handle)
```
