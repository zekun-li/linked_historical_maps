```python
import json
output = Path("output")

with open(output / 'url_osm_result.json') as f:
    url_dic = json.load(f)
    
with open(output / 'not_matching_phrases.json') as f:
    not_matching_dic3 = json.load(f)

with open('map_phrase_USGS(GT).json') as f:
    history = json.load(f)
```

```python
not_matching_dic3
```

```python
number_matching = dict()
for map_name in url_dic:
    number_matching[map_name] = {"matching": len(url_dic[map_name])}
for map_name in number_matching:
    number_matching[map_name].update({"total": len(history[map_name])})

for map_name in number_matching:
    number_matching[map_name].update({"not matching": len(not_matching_dic3[map_name])})
    
    if len(not_matching_dic3[map_name]) != len(history[map_name]) - len(url_dic[map_name].keys()):
        print(map_name)
```
