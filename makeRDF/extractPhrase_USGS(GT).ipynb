```python
"""
# extract phrase from ground truth
# USGS-15-CA-hesperia-e1902-s1898-rp1912: empty data, so not used

input: annotation csv files (ground truth) in "input" folder
output: map_phrase_USGS(GT).json {map: [phrases]}
"""

import glob
import pandas as pd

# read all input files and check column names
for filename in glob.glob("input/*.csv"):
    print(filename)
    print(pd.read_csv(filename).columns)
    print("========")
```

```python
# make phrase set per each map jpg file
# remove nan, lowercase
phrase_dic = dict()

for filename in glob.glob("input/*.csv"):
    key = filename[6:-3] +'jpg'
    try:
        try:
            phrase_dic[key] = list(set([ele.lower() for ele in set(pd.read_csv(filename).phrase) if str(ele) != 'nan']))
        except:
            phrase_dic[key] = list(set([ele.lower() for ele in set(pd.read_csv(filename).PHRASE) if str(ele) != 'nan']))
    except:
        try:
            phrase_dic[key] = list(set([ele.lower() for ele in set(pd.read_csv(filename).Phrase) if str(ele) != 'nan']))
        except:
            phrase_dic[key] = list(set([ele.lower() for ele in set(pd.read_csv(filename).Text) if str(ele) != 'nan']))
```

```python
import json

with open('map_phrase_USGS(GT).json', 'w') as f:
    json.dump(phrase_dic, f)
```
